```python
df["date_added"] = pd.to_datetime(df["date_added"], format="mixed")
df["day_added"] = df["date_added"].dt.day
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month
df["year_added"].astype(int)
df["day_added"].astype(int)
```

# Data Visualization

```python
sns.countplot(x=netflix["type"], hue=netflix["type"], width=0.5)
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x=netflix["rating"], hue=netflix["type"])
plt.show()
```

```python
labels = ["Movie", "TV show"]
size = netflix["type"].value_counts()
explode = [0, 0.1]
plt.pie(size, labels=netflix.type.unique(), explode=explode, shadow=True, startangle=90)
plt.title("Distribution of Type", fontsize=20)
plt.legend()
plt.show()
```
