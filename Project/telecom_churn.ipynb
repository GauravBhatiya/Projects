```python
replace = {"no": False, "yes": True}
df["international plan"] = df["international plan"].map(replace)
df.head()
```

```python
df.replace({"voice mail plan": replace}, inplace=True)
df.head()
```

# Grouping

```python
columns_to_show = ["total day minutes", "total eve minutes", "total night minutes"]

df.groupby(["churn"])[columns_to_show].describe(percentiles=[])
```

```python
columns_to_show = ["total day minutes", "total eve minutes", "total night minutes"]
df.groupby(["churn"])[columns_to_show].agg(["mean", "max", "std", "count"])
```

```python
df.pivot_table(
    ["total day calls", "total eve calls", "total night calls"],
    ["churn"],
    aggfunc="mean",
)
```

```python
sns.countplot(data=df, x="international plan", hue="churn")
plt.show() 
```

```python
sns.barplot(y="customer service calls", x="international plan", hue="churn", data=df)
```
